```python
import requests

def call_api(text, api_url="http://127.0.0.1:5000/segment-analysis"):
    # Gửi yêu cầu POST đến API
    payload = {'text': text}
    response = requests.post(api_url, json=payload)

    # Kiểm tra xem yêu cầu có thành công không (status code 200)
    if response.status_code == 200:
        result = response.json()
        return result
    else:
        # In thông báo lỗi nếu có lỗi
        print(f"Error: {response.status_code}")
        return None

# Sử dụng hàm để gửi đoạn văn bản và nhận kết quả
text_to_predict = "This is a sample text for prediction."
prediction_result = call_api(text_to_predict)

if prediction_result:
    print("Prediction result:", prediction_result)
else:
    print("Failed to get prediction result.")
```

```python
import httpx

def send_comment_to_api(comment):
    url = "http://127.0.0.1:5000/predict"  # Điều chỉnh URL nếu cần thiết
    data = {"text": comment}

    try:
        response = httpx.post(url, data=data)
        response.raise_for_status()  # Nếu có lỗi, raise exception
        result = response.json()
        return result['result'][0]['label']
    except Exception as e:
        return {"success": False, "error": str(e)}

# Sử dụng hàm để gửi comment và in ra kết quả
comment_to_send = "so deleciuos"
result = send_comment_to_api(comment_to_send)

```
